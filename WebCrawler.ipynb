```python
def get_text_from_sel(sel):     
    mylist = [] 
    try: 
        results = r.html.find(sel) 
        for result in results: 
            mytext = result.text 
            mylist.append(mytext) 
        return mylist 
    except: 
        return None
```

```python
import pandas as pd

count = 0
frames = []
while count <= 200:
    r = session.get(url)
    df = pd.DataFrame()
    for item in ['name','category', 'location', 'price']:
        se = pd.Series(get_text_from_sel(eval(item)))
        df[item] = se.values
    frames.append(df)
    button = r.html.find(nextButton)
    url = list(button[0].absolute_links)[0]
    count += 1
    print(count)
    
result = pd.concat(frames).drop_duplicates()
```
