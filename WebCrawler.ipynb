```python
from requests_html import HTMLSession
session = HTMLSession()

url = 'https://www.99.co/singapore/sale?listing_type=sale&query_coords=1.3039947%2C103.8298507&query_type=city&show_future_mrts=true&zoom=11&radius_max=1000&query_limit=radius&property_segments=residential'
```

```python
def get_text_from_sel(sel):     
    mylist = [] 
    try: 
        results = r.html.find(sel) 
        for result in results: 
            mytext = result.text 
            mylist.append(mytext) 
        return mylist 
    except: 
        return None
```

```python
%%time
import pandas as pd

count = 0
frames = []
while count <= 200:
    r = session.get(url)
    df = pd.DataFrame()
    for item in ['name','category', 'location','area_sqft', 'price']:
        vals = get_text_from_sel(eval(item))
        if (item == 'price'): vals = [int(val[1:].replace(',',''))if val!='NaN' else val for val in vals ]
        if (item == 'area_sqft'): vals = [int(val[:-4].replace(',','')) if val!='NaN'else val for val in vals]  
        se = pd.Series(vals)        
        df[item] = se.values
    frames.append(df)
    button = r.html.find(nextButton)
    url = list(button[0].absolute_links)[0]
    count += 1
    #print(count)
    
result = pd.concat(frames).drop_duplicates()
```
